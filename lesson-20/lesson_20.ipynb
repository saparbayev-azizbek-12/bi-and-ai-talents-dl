```python
from transformers import AutoModelForSequenceClassification

model_name = "sentence-transformers/all-MiniLM-L6-v2"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
exp1 = "This PC is good"
exp2 = "This PC is not bad"

tkn1 = tokenizer(exp1, return_tensors='pt')
tkn2 = tokenizer(exp2, return_tensors='pt')
```

```python
tkn1
```

```python
lg1 = model(**tkn1)
lg2 = model(**tkn2)
```

```python
import torch

prob1 = torch.argmax(torch.softmax(lg1.logits, dim=-1), dim=-1)
prob2 = torch.argmax(torch.softmax(lg2.logits, dim=-1), dim=-1)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_name = "google-bert/bert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
```

```python
from datasets import load_dataset

dataset = load_dataset("imdb")
```

```python
dataset["train"][0]
```

```python
def tokenize_fn(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=256
    )

tokenized_dataset = dataset.map(tokenize_fn, batched=True)
```

```python
tokenized_dataset["train"][0].keys()
```
