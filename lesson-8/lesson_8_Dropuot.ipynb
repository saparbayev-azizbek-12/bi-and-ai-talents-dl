```python
import torch
```

```python
class CustomDropout:
  def __init__(self, p=0.2):
    self.p = p
    self.training = True

  def train(self):
    self.training = True

  def eval(self):
    self.training = False

  def __call__(self, x):
    if not self.training:
      return
    mask = (torch.rand_like(x) > self.p).float()
    return mask
```
